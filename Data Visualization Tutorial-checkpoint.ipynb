```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Matlibplot vs seaborn

# Loading data


```python
df = pd.read_csv('../DATA/gapminder.csv')
display(df)
```

```python
us = df.loc [df.country == "United States", :]
us.head()
```

```python
# US life_expectancy

plt.plot(us.year, us.life_expectancy)
plt.title("US life expectancy over years")
plt.ylabel("Life Expectancy")
plt.xlabel("Years")
plt.grid()
plt.show()
```

```python
# Ban Charts
df_2007 = df. loc[df.year ==2007,:].sort_values('gdp_per_capita', ascending=False).head(10)
df_2007.head()

# Plotting
plt. figure(figsize=(4,3))
plt.bar (df_2007.country, df_2007.gdp_per_capita)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Scatter Plot
df_2007 = df. loc[df .year==2007]
plt. figure(figsize=(6,4))
plt. scatter(df_2007.gdp_per_capita, df_2007.life_expectancy, alpha=0.7) 
plt.show()
```

```python
# Histrogram Plot
df_2007 = df. loc[df .year==2007]
plt. figure(figsize=(6,4))
plt.hist(df_2007.life_expectancy,bins = 50, color="skyblue", edgecolor = "black") 
plt.show()
```

```python
# Pie Chart
popRegion = df_2007.groupby('region')['population'].sum()
plt.figure(figsize=(6,4))
plt.pie(popRegion, labels=popRegion.index) 
plt.show()
```
