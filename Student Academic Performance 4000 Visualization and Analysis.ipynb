```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#https://www.kaggle.com/datasets/firedmosquito831/student-academic-performance-simulation-4000?resource=download
```

```python
df = pd.read_csv("../DATA/SAP-4000.csv")
display(df)
```

```python

#Checking for NAs
df.isna().sum()
```

```python
#Group by male and female
gender_pop = df.groupby(by="Gender").size()
print(f"{gender_pop}")
```

```python
#Mean Study Hours
average_time_studying = df ["HoursStudied/Week"].mean()
print (f"{average_time_studying}")
```

```python
#Median Study Hours
median_time_studying = df ["HoursStudied/Week"].median()
print (f"{median_time_studying}")
```

```python
#Tutoring
df.groupby(by="Tutoring").size()
```

```python
#Region	Attendance(%)	Parent Education	Exam_Score

group_region = df.groupby(by="Region").size()
group_education = df.groupby(by="Parent Education").size()

print(f"{group_region} group_region")

print(f"{group_education} group_education")
```

```python
# Max and Min Exam Score

max_exam = df ["Exam_Score"].max()
print (f"{max_exam} max exam score")


mean_exam = df ["Exam_Score"].mean().round()
print (f"{mean_exam} mean exam score")

min_exam = df ["Exam_Score"].min()
print (f"{min_exam} min exam score")
```

```python
#Data Visualisation


# Count the number of Males and Females
df_gender = df['Gender'].value_counts()

# Check the counts
print(df_gender.head())

plt.figure(figsize=(4, 3))
plt.bar(df_gender.index, df_gender.values)
plt.xticks(rotation=90)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.show()
```

```python
# Calculate average hours studied per week, grouped by Parent Education
average_time_studying = df.groupby("Parent Education")["HoursStudied/Week"].mean()

# Plot
plt.figure(figsize=(6,4))
plt.xlabel("Parent Education")
plt.ylim([0,15])
plt.ylabel("Average Hours Studied/Week")
plt.title("Average Study Time by Parent Education")
plt.bar(average_time_studying.index, average_time_studying.values)
plt.xticks(rotation=45)
# Add value labels on top of each bar
for i, value in enumerate(average_time_studying.values):
    plt.text(i, value + 0.1, f"{value:.2f}", ha='center', va='bottom')

plt.show()

```

```python
#Data Visualisation


# Count the number of Males and Females
df_gender = df['Gender'].value_counts()

df_attendance = df['Attendance(%)'].mean()

# Calculate mean attendance per region
mean_attendance_by_gender = df.groupby('Gender')['Attendance(%)'].mean()


plt.figure(figsize=(4, 3))
plt.bar(df_gender.index, df_attendance.mean())
plt.xticks(rotation=90)
plt.xlabel('Gender')
plt.ylabel('df_attendance')
plt.title(' Attendance Gender Distribution')


for i, value in enumerate(mean_attendance_by_gender):
    plt.text(i, value + 0.1, f"{value:.2f}", ha='center', va='bottom')
    
plt.show()
```
